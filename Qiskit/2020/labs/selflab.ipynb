```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.measure([0,1],[0,1])
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram

circuit.measure(2,2)

simulator = Aer.get_backend('qasm_simulator')
circuit = transpile(circuit, simulator)
result = simulator.run(circuit, shots=1024).result().get_counts()
plot_histogram(result)
```
